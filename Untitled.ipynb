```python
# Step 1: Create a synthetic dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Project Name": [f"Project {i+1}" for i in range(50)],
    "Region": np.random.choice(["North", "South", "East", "West"], size=50),
    "Project Type": np.random.choice(["Residential", "Commercial", "Infrastructure"], size=50),
    "Cost (USD)": np.random.randint(50000, 500000, size=50),
    "Completion Time (Months)": np.random.randint(6, 36, size=50),
    "Materials Cost (USD)": np.random.randint(20000, 200000, size=50),
    "Labor Cost (USD)": np.random.randint(15000, 150000, size=50),
    "Overhead Costs (USD)": np.random.randint(5000, 50000, size=50),
}

# Convert to DataFrame
df = pd.DataFrame(data)

    
```

```python
# Step 2: Enhanced Data Cleaning

# 1. Check and Handle Missing Values
def check_missing_values(df):
    missing = df.isnull().sum()
    missing_percentage = (missing / len(df)) * 100
    print("Missing Values (%):\n", missing_percentage[missing_percentage > 0])
    return missing_percentage

# Display missing value report
missing_values = check_missing_values(df)

# Fill missing values (example: median for numerical columns)
df.fillna(df.median(), inplace=True)

# 2. Add a Total Cost Column
df["Total Cost"] = df["Materials Cost (USD)"] + df["Labor Cost (USD)"] + df["Overhead Costs (USD)"]

# 3. Remove Outliers Using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    cleaned_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    print(f"Rows removed due to outliers in {column}: {len(df) - len(cleaned_df)}")
    return cleaned_df

# Apply outlier removal on the 'Cost (USD)' column
df_cleaned = remove_outliers(df, "Cost (USD)")

# 4. Post-Cleaning Insights
print(f"Data after cleaning: {len(df_cleaned)} rows")
print("Summary Statistics:\n", df_cleaned.describe())

# Optional: Save cleaned data
df_cleaned.to_csv("cleaned_data.csv", index=False)
```

```python
# Step 3: Descriptive Statistics
print(df.describe())
```

```python

# Enhanced Visualization 1: Cost by Project Type
plt.figure(figsize=(10, 6))
sns.barplot(x="Project Type", y="Cost (USD)", data=df, estimator=np.mean, ci=None, palette="viridis")
plt.title("Average Cost by Project Type", fontsize=14, fontweight='bold')
plt.ylabel("Average Cost (USD)", fontsize=12)
plt.xlabel("Project Type", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Enhanced Visualization 2: Cost Breakdown
cost_breakdown = df[["Materials Cost (USD)", "Labor Cost (USD)", "Overhead Costs (USD)"]].sum()
plt.figure(figsize=(8, 8))
cost_breakdown.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    colors=sns.color_palette("pastel"),
    explode=[0.05, 0.05, 0.05],
    shadow=True,
)
plt.title("Cost Breakdown by Component", fontsize=14, fontweight='bold')
plt.ylabel("")  # Suppress ylabel for cleaner output
plt.show()

# Enhanced Visualization 3: Cost vs. Completion Time
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Completion Time (Months)", 
    y="Cost (USD)", 
    hue="Project Type", 
    data=df, 
    palette="Set2", 
    size="Cost (USD)", 
    sizes=(20, 200), 
    alpha=0.7,
)
plt.title("Cost vs. Completion Time by Project Type", fontsize=14, fontweight='bold')
plt.xlabel("Completion Time (Months)", fontsize=12)
plt.ylabel("Cost (USD)", fontsize=12)
plt.legend(title="Project Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Step 5: Enhanced Cost Optimization Analysis

# Analyze average cost components
cost_components = df[["Materials Cost (USD)", "Labor Cost (USD)", "Overhead Costs (USD)"]].mean()
print("Average Cost Components (USD):")
print(cost_components)

# Calculate percentage contribution of each component
total_mean_cost = cost_components.sum()
cost_percentage = (cost_components / total_mean_cost) * 100
print("\nCost Component Contribution (%):")
print(cost_percentage)

# Visualization: Cost Component Contributions
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
cost_components.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'], alpha=0.8)
plt.title("Average Cost Components")
plt.ylabel("Cost (USD)")
plt.xlabel("Cost Component")
plt.xticks(rotation=45)
plt.show()

# Pie chart for percentage contribution
plt.figure(figsize=(8, 8))
cost_percentage.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title("Cost Component Contribution (%)")
plt.ylabel("")  # Hide ylabel for a cleaner look
plt.show()

# Identify key drivers: Highlight components exceeding 40% contribution
high_contributors = cost_percentage[cost_percentage > 40].index.tolist()
print("\nKey Cost Drivers (exceeding 40% contribution):")
print(high_contributors)

# Dynamic Suggestions
print("\nSuggestions:")
if "Materials Cost (USD)" in high_contributors:
    print("Consider exploring alternative suppliers or using recycled materials to reduce material costs.")
if "Labor Cost (USD)" in high_contributors:
    print("Evaluate labor allocation and negotiate rates in regions with higher labor costs.")
if "Overhead Costs (USD)" in high_contributors:
    print("Optimize operational processes to reduce overhead costs (e.g., energy, equipment maintenance).")
```
